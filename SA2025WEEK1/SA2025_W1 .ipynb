### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
# Calculate median mpg
median_mpg = df['mpg'].median()

# Filter cars with above-median mpg
above_median_mpg = df[df['mpg'] > median_mpg]

# Calculate horsepower-to-weight ratio
above_median_mpg['hp_weight_ratio'] = above_median_mpg['horsepower'] / above_median_mpg['weight']

# Find the max ratio
max_ratio = above_median_mpg['hp_weight_ratio'].max()

# Return the car(s) with this ratio
best_cars = above_median_mpg[above_median_mpg['hp_weight_ratio'] == max_ratio][['name', 'horsepower', 'weight', 'hp_weight_ratio']]
print(best_cars)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
import seaborn as sns
import matplotlib.pyplot as plt

# Group and compute average mpg
mpg_trend = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=mpg_trend, x='model_year', y='mpg', hue='origin', marker='o')
plt.title('Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
# Group by name and check which ones appear in multiple model years
car_years = df.groupby('name')['model_year'].nunique()
repeated_models = car_years[car_years > 1].index

# Filter data for those models
multi_year_models = df[df['name'].isin(repeated_models)]

# Group by car name and compute std and count of mpg
consistency_stats = multi_year_models.groupby('name')['mpg'].agg(['std', 'count', 'mean']).reset_index()
consistent_cars = consistency_stats[(consistency_stats['std'] < 1.0) & (consistency_stats['count'] >= 2)]

# Sort by count (descending) then mean mpg (descending)
consistent_cars = consistent_cars.sort_values(by=['count', 'mean'], ascending=[False, False])
consistent_cars.rename(columns={'count': 'appearances', 'mean': 'average_mpg'}, inplace=True)

print(consistent_cars[['name', 'appearances', 'average_mpg']])
```
